```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
students_performance = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/prework_data/main/students_performance.csv")
```

```python
students_performance.head()
```

```python
students_performance.shape
```

```python
students_performance.nunique() #utilizamos esta función para ver el número de distintos valores
```

```python
sns.histplot(x="gender", data=students_performance)
```

```python
sns.histplot(x="math score", data=students_performance) #la gráfica es un histograma, empleado principalmente en valores numéricos, aunque también podría ser útil para visualizar datos de otro modo
```

```python
sns.countplot(x="gender", data=students_performance)
```

```python
sns.countplot(x="math score", data=students_performance) #podemos observar que al ver muchos valores únicos, la gráfica se estrecha y se hace poco legible, no obstante, puede ser muy útil para observar, por ejemplo, tendencias
```

```python
students_performance["math score"].nunique() #como podemos comprobar hay 81 valores distintos, de ahí la mayor complejidad de la gráfica
```

```python
students_performance["math score"].unique()  #estos son todos los valores que nos aparecería en el axis 1 u horizontal
```

```python
sns.histplot(x="math score", data=students_performance)
```

```python
#al observar de nuevo esta gráfica, podemos a simple vista concluir que los resultados en general son positivos, siendo el mínimo 0 y el máximo 100. La mayoría se mueve entre una puntuación de 60 a 80. Podemos también observar que los resultados altos son mucho mayores que los bajos. Al igual que con las funciones, podemos obtener datos muy relevantes como la media, max, min, moda; esta gráfica otorga el mismo conocimiento sin proporcionar valores exactos. Podría ser ideal para una previsualización antes de empezar un proyecto.
```
